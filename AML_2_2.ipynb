```
import math
import csv
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```
with open('/content/drive/My Drive/seen/seen-dataset/dataset_seen_training_siamese.csv', 'r') as f:
    reader = csv.reader(f)
    count=0
    t=[]
    cmatrix = []
    smatrix= []
    for row in reader:
        if (count!=0):
            data=[]
            data1=[]
            with open('/content/drive/My Drive/15features.csv','r') as f1:
                reader1 = csv.reader(f1)
                count1=0
                flag1=0
                for row1 in reader1:
                    if (count1!=0):
                        if (row1[0]==row[1]):
                            for column in row1:
                                data.append(column)
                                flag1=1
                            break
                    count1=count1+1
                data=data[1:]
                data=[int(data) for data in data]
            count2=0
            with open('/content/drive/My Drive/15features.csv','r') as f1:
                reader1 = csv.reader(f1)
                flag2=0
                for row1 in reader1:
                    if (count2!=0):
                        if (row1[0]==row[2]):
                            for column in row1:
                                data1.append(column)
                                flag2=1
                            break
                    count2=count2+1
                data1=data1[1:]
                data1=[int(data1) for data1 in data1]
                concatenate=data+data1
                sub=[abs((data)-(data1)) for data,data1 in zip(data,data1)]
                if(flag1==1 & flag2==1):
                  cmatrix.append(concatenate) 
                  smatrix.append(sub)
                  t.append(int(row[3]))
        count=count+1   
```

```
with open('/content/drive/My Drive/seen/seen-dataset/dataset_seen_validation_siamese.csv', 'r') as f:
    reader = csv.reader(f)
    count=0
    tval=[]
    cval = []
    sval= []
    for row in reader:
        if (count!=0):
            data=[]
            data1=[]
            with open('/content/drive/My Drive/15features.csv','r') as f1:
                reader1 = csv.reader(f1)
                count1=0
                flag1=0
                for row1 in reader1:
                    if (count1!=0):
                        if (row1[0]==row[1]):
                            for column in row1:
                                data.append(column)
                                flag1=1
                            break
                    count1=count1+1
                data=data[1:]
                data=[int(data) for data in data]
            count2=0
            with open('/content/drive/My Drive/15features.csv','r') as f1:
                reader1 = csv.reader(f1)
                flag2=0
                for row1 in reader1:
                    if (count2!=0):
                        if (row1[0]==row[2]):
                            for column in row1:
                                data1.append(column)
                                flag2=1
                            break
                    count2=count2+1
                data1=data1[1:]
                data1=[int(data1) for data1 in data1]
                concatenate=data+data1
                sub=[abs((data)-(data1)) for data,data1 in zip(data,data1)]
                if(flag1==1 & flag2==1):
                  cval.append(concatenate) 
                  sval.append(sub)
                  tval.append(int(row[3]))
        count=count+1   
```

```
cmat=np.matrix(cmatrix)
smat=np.matrix(smatrix)
target=np.matrix(t)
cmat.shape
```

```
cval=np.matrix(cval)
sval=np.matrix(sval)
sval_df=pd.DataFrame(sval)
sval_df.columns=['f1', 'f2', 'f3','f4', 'f5', 'f6','f7', 'f8', 'f9','f10', 'f11', 'f12','f13', 'f14', 'f15']
cval_df=pd.DataFrame(cval)
cval_df.columns=['f1', 'f2', 'f3','f4', 'f5', 'f6','f7', 'f8', 'f9','f10', 'f11', 'f12','f13', 'f14', 'f15',
               'g1', 'g2', 'g3','g4', 'g5', 'g6','g7', 'g8', 'g9','g10', 'g11', 'g12','g13', 'g14', 'g15']
```
